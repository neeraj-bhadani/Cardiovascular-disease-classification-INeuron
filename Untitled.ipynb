```python
import pandas as pd
import numpy as np
pd.set_option('display.float_format', lambda x: '{:.3f}'.format(x))
pd.set_option('display.max_columns',None)
```

```python
data = pd.read_csv('cleaned_data')
data.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
data.head()
```

```python
num_features = ['age','height','weight','ap_hi','ap_lo']
cat_features =['gender','cholestrol','gluc','smoke','alco','active']
```

```python
nominal_cat = ['gender','smoke','alco','active']
ordinal_cat = ['cholesterol','gluc']
```

```python
data = pd.get_dummies(data, columns = nominal_cat)
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
data =  pd.DataFrame(scale.fit_transform(data),columns=data.columns)
data
```

```python
y = data['cardio']
y.head()
```

```python
X=data.copy()
X.drop('cardio',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier 
clf = RandomForestClassifier(max_depth=3)
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', C=1).fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, \
    recall_score, confusion_matrix, classification_report, \
    accuracy_score, f1_score
```

```python
print('Accuracy:', accuracy_score(y_test, prediction))
print('F1 score:', f1_score(y_test, prediction))
print('Recall:', recall_score(y_test, prediction))
print('Precision:', precision_score(y_test, prediction))
print('\n clasification report:\n', classification_report(y_test,prediction))
print('\n confussion matrix:\n',confusion_matrix(y_test, prediction))

```
